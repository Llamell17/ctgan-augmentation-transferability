## Clustering

Original text: the collected data is sequentially divided (suffle=False) into training and test sets at an 8:2 ration.

```python
# Using just geological conditions as variables
X_km = df[['UCS', 'RD']].copy()
X_km_train, X_km_test = train_test_split(X_km, test_size=0.2, shuffle=False)

print("Training set has {} samples.".format(X_km_train.shape[0]))
print("Testing set has {} samples.".format(X_km_test.shape[0]))
```

Kmeans: with cluster numbers ranging from 2 to 5. 

```python
range_n = range(2, 6)
scores = []
cluster_results = {}

for n in range_n:
    km = KMeans(n_clusters=n, random_state=42)
    km.fit(X_km_train)
    clusters = km.labels_
    centroids = km.cluster_centers_

    # Silhouette
    score = silhouette_score(X_km_train, clusters)
    scores.append(score)
    print(f"Silhouette Coefficient, {n} clusters: {score:.3f}")
    
    label = 'clusters_' + str(n)
    cluster_results[label] = clusters

    # Plot
    plt.figure(figsize=(5, 5))
    plt.scatter(X_km_train["UCS"], X_km_train["RD"], c=clusters, s=60, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=250, color='red', label="Centroids")
    plt.title(f"Silhouette Coefficient = {score:.3f}   (n={n})")
    plt.xlabel("Uniaxial compressive strength")
    plt.xlim(31, 50) #32.6, 49.1
    plt.ylabel("Rock density")
    plt.ylim(2.25, 2.9) #2.3, 2.8
    plt.legend()
    plt.show()
```

```python
# Plot silhouette scores
plt.figure(figsize=(4, 4))
plt.plot(list(range_n), scores, marker='o')
plt.xlabel('n')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis for K‑Means')
plt.show()
```

Silhouette coeffiicient of 0.707 with n=2: 69 and 21 each cathegory. (76.66%), 23 de test

69: predominance of limestone >> UCS RD

```python
c0_0, c1_0 = 69 , 21
proportion_0 = c0_0/ (c0_0 + c1_0)
print('Original:', proportion_0)

x = 250
c0_1, c1_1 = 69 + x, 21 + x
proportion_1 = c0_1/ (c0_1 + c1_1)
print('Hybrid:',proportion_1)
```

```python
# Keep the best cluster and delete the others
best_clusters = cluster_results['clusters_2']
df.loc[X_km_train.index, "cluster"] = best_clusters

df_c0_train = df[df["cluster"] == 0].drop(columns=["cluster"])
df_c0_train.to_csv("0_cluster0.csv", index=False)

df_c1_train = df[df["cluster"] == 1].drop(columns=["cluster"])
df_c1_train.to_csv("0_cluster1.csv", index=False)

df_test = df.loc[X_km_test.index]
df_test.to_csv("0_test.csv", index=False)

n_c0_0 = len(df_c0_train)
n_c1_0 = len(df_c1_train)
print("Limestone represents {}%.".format(n_c0_0/(n_c0_0+n_c1_0)*100))
```
